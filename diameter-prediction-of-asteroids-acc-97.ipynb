```python
def handle_outliers_iqr(df, columns=None, method='remove'):

    df_out = df.copy()

    if columns is None:
        columns = df_out.select_dtypes(include=['number']).columns.tolist()

    for col in columns:
        Q1 = df_out[col].quantile(0.25)
        Q3 = df_out[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if method == 'remove':
            df_out = df_out[(df_out[col] >= lower_bound) & (df_out[col] <= upper_bound)]

        elif method == 'cap':
            df_out[col] = df_out[col].clip(lower=lower_bound, upper=upper_bound)

        else:
            raise ValueError("method must be either 'remove' or 'cap'")

    return df_out
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Lasso, Ridge
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

models = {
    'Lasso Regression': make_pipeline(
        PolynomialFeatures(degree=2, include_bias=False),
        Lasso(alpha=0.1, random_state=42)
    ),
    'Ridge Regression': make_pipeline(
        PolynomialFeatures(degree=2, include_bias=False),
        Ridge(alpha=1.0, random_state=42)
    )
}

for name, model in models.items():
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    
    print(f"=== {name} ===")
    print(f"Train MSE: {train_mse:.4f}, MAE: {train_mae:.4f}")
    print(f"Test  MSE: {test_mse:.4f}, MAE: {test_mae:.4f}")
    print(f"R2 Train: {train_r2:.4f}")
    print(f"R2 test: {test_r2:.4f}\n")
```
